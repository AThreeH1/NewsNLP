```python
#Importing Libraries
import os
import pandas as pd
```

```python
# Alloting paths
PathP = r"C:\Study\Finance\IIMV Intern\News_posutf.csv"
PathN = r"C:\Study\Finance\IIMV Intern\News_negutf.csv"

# Reading files
Positive = pd.read_csv(PathP)
Negative = pd.read_csv(PathN)

# Displaying first 5 rows
print(Positive.head(5))
print(Negative.head(5))
```

```python
#removing NAN
pdf = Positive.dropna(subset=['Company'], inplace = False)
ndf = Negative.dropna(subset=['Company'], inplace = False)

ndf.head(-1)
```

```python
#creating new columns to represent positive/negative
pdf.loc[:, 'Type'] = 1
ndf.loc[:, 'Type'] = 0

#adding the dataframes together 
df = pd.concat([pdf,ndf], axis = 0, ignore_index = True)

#re-check
df.info()
```

```python
#checking the entries
df.head(5)
df.tail(5)
```

```python
#setting directory
Out = "C:\Study\Finance\IIMV Intern\\"

#copying req column
Companies = df['Company'].copy()

#saving files
output = os.path.join(Out, 'companies.csv')
Companies.to_csv(output, index = False)
```
